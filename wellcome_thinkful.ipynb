```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import scipy.stats
import seaborn as sns
%matplotlib inline
df=pd.read_csv('WELLCOME_APCspend2013_forThinkful.csv', encoding = "ISO-8859-1")
```

```python
#To remove duplicates by capitalization or whitespace
df['Journal title']=df['Journal title'].str.lower()
df['Journal title']=df['Journal title'].str.strip()
```

```python
df['Journal title']=df['Journal title'].str.replace(',', ':', regex=False)
```

```python
#Used groupby Journal title to indentify the unique instances of journals not matching
#fixing journal issues
df['Journal title']=df['Journal title'].str.replace('j med chem', 'journal of medicinal chemistry', regex=False)
df['Journal title']=df['Journal title'].str.replace('acta crystallographica: section d', 'acta crystallographica section d: biological crystallography', regex=False)
df['Journal title']=df['Journal title'].str.replace('acta crystallography d', 'acta crystallographica section d: biological crystallography', regex=False)
df['Journal title']=df['Journal title'].str.replace('acta d', 'acta crystallographica section d: biological crystallography', regex=False)
df['Journal title']=df['Journal title'].str.replace('acta crystallographica section d: biological crystallographyermato venereologica', 'acta crystallographica section d: biological crystallography', regex=False)
df['Journal title']=df['Journal title'].str.replace('acta f', 'acta crystallographica section f: structural biology and crystallization communications', regex=False)
df['Journal title']=df['Journal title'].str.replace('trop med int health', 'tropical medicine and international health', regex=False)
df['Journal title']=df['Journal title'].str.replace('plosone', 'plos one', regex=False)
#Note virology and virology journal are infact two seperate journals
```
